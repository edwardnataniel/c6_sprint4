```python
import wordcloud
from bs4 import BeautifulSoup
```

```python
with open("your_posts_1.html", encoding="utf8") as page:
    soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

```python
contents = soup.find_all("div", class_="_2pin")
```

```python
contents
```

```python
# [i.get_text() for i in contents]
```

```python
x = []
for i in contents:
    if "Updated" not in i.get_text():
        x.append(i.get_text())
```

```python
x
```

```python
y = ' '.join(x)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60,100)

from PIL import Image
import numpy as np

mask = np.array(Image.open("o.png"))

wordcloud = WordCloud(background_color="black", max_words=2000, width=1024, height=720, mask=mask)

wordcloud.generate(y)
plt.figure(figsize=(20,12))
plt.title("Facebook Wordcloud")
plt.imshow(wordcloud.recolor(color_func=grey_color_func, random_state=3), interpolation='bilinear')
plt.axis("off")
plt.show()
```
