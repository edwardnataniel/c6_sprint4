```python
import pandas as pd
import numpy as np
```

```python
hot_jokes = pd.read_csv('hot_jokes.csv')
top_jokes = pd.read_csv('top_jokes.csv')
```

```python
top_jokes.shape
```

```python
hot_jokes.shape
```

```python
hot_jokes = hot_jokes.drop(['stickied','Unnamed: 0'], axis=1)
top_jokes = top_jokes.drop(['stickied','Unnamed: 0'], axis=1)
```

```python
hot_jokes['hot'] = True
top_jokes['top'] = True
```

```python
merged_jokes = pd.concat([hot_jokes, top_jokes])
```

```python
merged_jokes
```

```python
merged_jokes['hot'] = merged_jokes['hot'].map({True:1, np.nan:0})
merged_jokes['top'] = merged_jokes['top'].map({True:1, np.nan:0})
```

```python
merged_jokes
```

```python
merged_jokes.to_csv('merged_jokes.csv')
```
